```python
minor_tag_dictionary = {
    'جهان': 3,
    'ایران': 2,
    'سیاسی > دولت': 4,
    'سیاسی > سیاست خارجی': 4,
    'اجتماعی > سلامت': 0,
    'آلمان': 3,
    'سیاسی > مجلس': 4,
    'سیاسی > سیاست داخلی': 4,
    'دانش و محیط زیست': 5,
    'سیاسی > حقوقی و قضایی': 4,
    'اقتصاد': 1,
    'سیاسی > دفاعی - امنيتی': 4,
    'اقتصادی > اقتصاد کلان': 1,
    'اجتماعی > حوادث، انتظامی': 0,
    'فرهنگ و هنر': 5,
    'ورزشی > فوتبال، فوتسال': 5,
    'علمی و دانشگاهی > آموزش': 5,
    'اقتصادی > تولید و تجارت': 1,
    'اجتماعی > جامعه، شهری': 0,
    'اقتصادی > عمران و اشتغال': 1,
    'بین الملل > آسیا،خاورمیانه': 3,
    'اجتماعی > آموزش و پرورش': 0,
    'فرهنگی و هنری > گردشگری و میراث': 5
}

data_train_set['tags'] = data_train_set['tags'].map(minor_tag_dictionary)
data_train_set['tags'] = pd.to_numeric(data_train_set['tags'], errors='coerce')
data_train_set = data_train_set.dropna(subset=['tags'])
```

```python
#Data Pre-Processing

import re


def simple_pre_process_data(text):
    # Remove Hyperlinks
    text = re.sub(r'https?:\/\/.*[\r\n]*', '', text)
    # Remove Arbitrary characters
    text = re.sub(r'#', '', text)
    # Remove Numbers
    text = re.sub(r'\d+', '', text)
    return text
```

```python
from hazm import Normalizer, Stemmer, Lemmatizer, stopwords_list,word_tokenize
from string import punctuation, printable, digits

normalizer = Normalizer()
stemmer = Stemmer()
stop_words = stopwords_list()
lemmatizer = Lemmatizer()

def advanced_pre_process_data(text):
    text = simple_pre_process_data(text)
    # Normalization
    text = normalizer.normalize(text)
    # Split Texts into Tokens
    token_list = []
    text_tokens = word_tokenize(text)
    for word in text_tokens:
        if word not in stop_words and word not in punctuation and word not in printable and word not in digits:
        # Stemming
            word = stemmer.stem(word)
        # Lemmatization
            word =  lemmatizer.lemmatize(word)
            token_list.append(word)
    return ' '.join(token_list)

data_train_set['title'] = data_train_set['title'].apply(advanced_pre_process_data)
data_train_set['description'] = data_train_set['description'].apply(advanced_pre_process_data)
```

```python
# Split Training and Testing Data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_train_set['title'] + ' ' + data_train_set['description'], data_train_set['tags'], test_size=0.2, random_state=42)
print('training data shape:', X_train.shape)
print('testing data shape: ', X_test.shape)
```

```python
# Text to Vector

from gensim.models import Word2Vec

sentences = [sentence.split() for sentence in X_train]
w2v_model = Word2Vec(sentences, window=5, min_count=5, workers=4)

def vectorize(sentence):
    words = sentence.split()
    words_vecs = [w2v_model.wv[word] for word in words if word in w2v_model.wv]
    if len(words_vecs) == 0:
        return np.zeros(100)
    words_vecs = np.array(words_vecs)
    return words_vecs.mean(axis=0)

X_train = np.array([vectorize(sentence) for sentence in X_train])
X_test = np.array([vectorize(sentence) for sentence in X_test])
```

```python
#Model Implementation

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True]
}

rfc = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=3, n_jobs=-1, verbose=3)
grid_search.fit(X_train, y_train)
best_rfc = grid_search.best_estimator_
best_rfc.fit(X_train, y_train)
y_pred = best_rfc.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
rf = RandomForestClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)
svc = SVC(probability=True, random_state=42)

# Define the parameter grids with efficient parameters
param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [None, 5],
    'min_samples_split': [2, 5]
}

param_grid_gb = {
    'n_estimators': [50],
    'learning_rate': [0.05, 0.1],
    'max_depth': [5]
}

param_grid_svc = {
    'C': [0.1, 1],
    'kernel': ['rbf']
}


gs_rf = GridSearchCV(rf, param_grid_rf, cv=5, n_jobs=-1, scoring='accuracy',verbose=3)
gs_rf.fit(X_train, y_train)
best_rf = gs_rf.best_estimator_

gs_gb = GridSearchCV(gb, param_grid_gb, cv=5, n_jobs=-1, scoring='accuracy',verbose=3)
gs_gb.fit(X_train, y_train)
best_gb = gs_gb.best_estimator_

gs_svc = GridSearchCV(svc, param_grid_svc, cv=5, n_jobs=-1, scoring='accuracy',verbose=3)
gs_svc.fit(X_train, y_train)
best_svc = gs_svc.best_estimator_

voting_clf = VotingClassifier(estimators=[('rf', best_rf), ('gb', best_gb), ('svc', best_svc)], voting='soft')

voting_clf.fit(X_train, y_train)

y_pred = voting_clf.predict(X_test)


print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
data_test_set['title'] = data_test_set['title'].apply(advanced_pre_process_data)
data_test_set['description'] = data_test_set['description'].apply(advanced_pre_process_data)
Xtest = data_test_set['title'] + ' ' + data_test_set['description']
Xtest = np.array([vectorize(sentence) for sentence in Xtest])
data_test_set['prediction'] = best_rfc.predict(Xtest)
result_data_main = data_test_set['prediction'].astype(int)
result_data_main.to_csv('submission.csv', index=False)
```

```python
#MLP

from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.layers import Dropout

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

model = Sequential()

model.add(Dense(512, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(6, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train_cat, epochs=15, batch_size=32, validation_split=0.2)

test_loss, test_accuracy = model.evaluate(X_test, y_test_cat)
print(test_accuracy)
```
